```python
import pandas as pd

df = pd.read_csv("datasets/Datafiniti_Hotel_Reviews.csv")
df.head()
```

```python
df = df[['categories','city','name','reviews.rating','reviews.text','reviews.title','reviews.userCity']]
```

```python
df =df.rename(columns={
    'reviews.rating': "rating",
    'reviews.text': "text",
    'reviews.title': "title",
    'reviews.userCity': "userCity"

})
```

```python
# Elegimos la muestra (analizar entre 1500 y 3000)
#data = df.sample(1500, random_state=786).reset_index(drop=True)
data = df.sample(1500, random_state=786).reset_index(drop=True)
data.shape

```

```python
stop_words= ['make','room', 'hotel', 'stay', 'good', 'bad', 'und', 'der', 'great', 'go', 'lot', 'staff', 'get', 'would', 'night']

exp_nlp101 = setup(data = data, target = 'text', session_id = 1213, verbose=True, custom_stopwords=stop_words)
```

<a id = "1.0"></a>
<p style="font-size:20px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px;">
<strong>Entrenar LDA con 4 tópicos</strong></p>

```python
lda = create_model(model='lda', num_topics=3, verbose=True)
```

```python
print(lda)
```

<a id = "1.0"></a>

<strong>Asignar resultado</strong></p>

```python
lda_results = assign_model(lda)
```

```python
print(lda_results[lda_results['Dominant_Topic'] == 'Topic 0']['rating'].mean())
print(lda_results[lda_results['Dominant_Topic'] == 'Topic 1']['rating'].mean())
print(lda_results[lda_results['Dominant_Topic'] == 'Topic 2']['rating'].mean())
```

Durante el proceso de NLP se realizaron pruebas con los siguientes grupos de valores:
- Cantidad de tópicos: [2,3,4,5]
- Tamaño de la muestra: [1000, 1200, 1500, 2000, 3000]
- Stop words: ['make','room', 'hotel', 'stay', 'good', 'bad', 'und', 'der', 'great', 'go', 'lot', 'staff', 'get', 'would', 'night'] 

En cada iteración se fueron agregando stopwords, así como modificando los valores mencionados para buscar tópicos a los que se les pueda asignar algún sentido, así como una buena distribución de las filas en los mismos.

Se encontró el mejor resultado, el más factible de análisis, usando los valores:
- Cantidad de tópicos: 3
- Tamaño de la muestra: 1500
- Stop words: ['make','room', 'hotel', 'stay', 'good', 'bad', 'und', 'der', 'great', 'go', 'lot', 'staff', 'get', 'would', 'night'] 

Se le asigna el siguiente significado a cada tópico obtenido:
Tópico 1: Contiene principalmente las reseñas positivas. Esto se ve claramente al observar los gráficos de frecuencia de palabras y bigramas, como así también la nube de palabras. Podemos encontrar palabras como clean, comfortable, nice, helpful, friendly, etc. Además, si observamos el promedio del rating es de más de 4/5.

Tópico 0: Este tópico contiene reseñas mayormente negativas, pero que usan palabras constructivas y se refieren a cosas que se podrían mejorar. 

Topico 2: Aquí se concentran las reseñas definitivamente negativas, con un vocabulario más fuerte que en el tópico anterior, lo cual indicaría que quien escribió la reseña lo hizo con enojo. Se observan palabras como stain, terrible, dirty, disgusting, cigarette, critter y cockroach.

Analizando los promedios de ratings de cada topico, vemos que el Topico 1 es claramente el que tiene reviews buenas, mientras que los otros representan  malas. Sin embargo, el Topico 0 se aproxima al rating medio de 3, y el Topico 2 tiene valores mas negativos. El modelo bien seprara 2 tipos de reviews "malas" como las criticas constructivas y los descargos por enojo.
